# Introduction to Scikit Learn

## 0. end-to-end Scikit Learn Workflow

```python
# 1. Get the data ready
import pandas as pd
import numpy as np

hd = pd.read_csv("heart-disease.csv")
hd
```

```python
# create x features matrix
x = hd.drop("target", axis=1)
# create y labels
y = hd['target']
```

```python
# 2. Choos the right model and Hyperparameters

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
# We'll keep the default hyperpaprameters
clf.get_params()
```

```python
# 3.Fit the model to the training data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Trying fit the model and try to learn the patterns in the training data
clf.fit(x_train, y_train)
```

```python
#make prediction
y_pred = clf.predict(x_test)
y_pred
```

```python
x_test
```

```python
# 4.Evaluate the model
clf.score(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# 5. Improve model
np.random.seed(42)
for i in range(10,100,10):
    print(f"trying model with {i} estimators")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"Model accuracy : {clf.score(x_test, y_test) * 100:.2f}")
    print(" ")
```

```python
# 6. Save model and load it 
import pickle

pickle.dump(clf, open("randomforest1.pkl", "wb"))
```

```python
lm = pickle.load(open("randomforest1.pkl", "rb"))
lm.score(x_test, y_test)
```
