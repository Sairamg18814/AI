```python
x = [1,2,3,4]
y = [11,12,31,34]
plt.plot(x,y);
```

```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot() # adds axis
plt.show()
```

```python
# 2nd method
fig = plt.figure() #creates a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
#3rd method (recommended)

fig , ax = plt.subplots()
ax.plot(x, [50,100,200,250]);
```
